# Prediction of upcoming cases using Linear Regression

```python
data = df.set_index('Unnamed: 0')
data
```

```python
# First need to get the Hospitalized case on the daily basis.
day = data[data['Current Status']=='Hospitalized'].groupby(['Month' , 'Date'])[['Num Cases']].sum()
day
```

```python
### Checking how many day's data do we have in our Dataframe in the form of array

x = np.arange(len(day))
x
```

```python
## Now get the value of 'Y'
```

```python
y = day.values
y
```

```python
plt.figure(figsize = (10, 5))
plt.scatter(x,y)
plt.show()
```

### Now we will Train the data.

##### **For linear Regression , 'X' should always be in '2Dimensional' , here 'X' is in one dimension i.e in array therefore we will reshape the 'X'**

```python
#-1 because i dont want the number of rows to define,with -1 it will automatically decides what no of rows to be present,
# and , cols defined to be as 1 only, which is 2D
x = x.reshape(-1 , 1)
x
```

### For linear regession we need to use 'Scikit learn' library.

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

#### **With fit method we train the model with our dataset which is in x and y,which at the back end calculate its slope and everything**, ie calculate the ***y = mx+c , and the square of all losses.***

```python
reg.fit(x ,  y)
```

```python
y[364]
```

```python
pred = reg.predict([[365]])
print('Predicted cases for the day 365 are : ',pred)
```

```python
## will see what machine will predict for day zero , as we know what it predicts for y
y_pred=reg.predict(x)
y_pred
```

### To see how the best fit line will see in the graph with this dataset

```python
plt.scatter(x, y)
plt.plot(x , y_pred)
plt.show()
```

```python
score = reg.score(x , y)*100
print('This score function will give us the percent of how close our prediction is , and which is: ' ,score)
```

```python
inter = reg.intercept_
print('This will give us the point where the intercetion is met at: ',inter)
```

```python
coef = reg.coef_
print('The co-efficient for the data is :' , coef)
```

```python
next_pred = reg.predict([[400]])
print('The prediction for the near future days : ' , next_pred)
```
